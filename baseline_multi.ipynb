```python
def rerange_cols(a, permutation):
    idx = np.empty_like(permutation)
    idx[permutation] = np.arange(len(permutation))
    return a[idx, :]  # in-place modification of a
```

```python
# представим что у нас уже массив индексов
# shuffeld_kspace[schem_int, :]
shuffeld_kspace_rearrange_rows_multi = np.array([rerange_cols(shuffeld_kspace_multi[i], schem_uniq) for i in range(shuffeld_kspace_multi.shape[0])]) 
shuffeld_kspace_rearrange_rows_multi
```
