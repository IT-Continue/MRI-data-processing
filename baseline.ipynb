```python
# https://stackoverflow.com/questions/20265229/rearrange-columns-of-numpy-2d-array
# это если у нас указано в какой по индексу элемент переходит данный
def rerange_rows(a, permutation):
    idx = np.empty_like(permutation)
    idx[permutation] = np.arange(len(permutation))
    a[:, idx]  # return a rearranged copy
    return a[:, idx]  # in-place modification of a
```

```python
shuffeld_kspace.shape
```

```python
# представим что у нас уже массив индексов
shuffeld_kspace_rearrange_rows = rerange_rows(shuffeld_kspace, schem_int) # shuffeld_kspace[schem_int, :]
shuffeld_kspace_rearrange_rows
```

```python
# inverse Fourier Transform to get zero filled solution
import fastmri
from fastmri.data import transforms
masked_kspace = transforms.to_tensor(shuffeld_kspace_rearrange_rows[()])
masked_kspace
```

```python
image = fastmri.ifft2c(masked_kspace)
image
```

```python
# absolute value
image = fastmri.complex_abs(image)
image
```

```python
reconstructions = {'image': image}
```
